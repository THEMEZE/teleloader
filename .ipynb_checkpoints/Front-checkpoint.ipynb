```python
!pip install python-telegram-bot selenium


from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select
import time

# Initialisation du driver
driver = webdriver.Chrome()

# Ouvrir l'URL
driver.get("https://y2down.cc/fr/")  # Remplacez avec l'URL r√©elle

# Attendre que l'√©l√©ment input soit visible avant d'interagir
input_element = WebDriverWait(driver, 10).until(
    EC.visibility_of_element_located((By.ID, "link"))
)

# Interagir avec l'√©l√©ment input pour entrer l'URL
input_element.send_keys("https://www.youtube.com/watch?v=r_aojYqud_0")

# Attendre que l'√©l√©ment select soit visible
select_element = WebDriverWait(driver, 10).until(
    EC.visibility_of_element_located((By.ID, "format"))
)

# Cr√©er un objet Select pour interagir avec le <select>
select = Select(select_element)

# S√©lectionner l'option "4k" (ou tout autre format selon le cas)
select.select_by_value("4k")  # S√©lectionne l'option "WEBM (4K)"

# S√©lectionner l'option "1080" comme exemple suppl√©mentaire
time.sleep(1)  # Attendre 1 seconde pour simuler une interaction avant de changer
select.select_by_value("1080")  # S√©lectionne "MP4 (1080p)"
time.sleep(1)

# Trouver le bouton "Download" par son ID
download_button = driver.find_element(By.ID, "load")

# Cliquer sur le bouton "Download" pour initier le processus
download_button.click()

# Augmenter le d√©lai d'attente pour donner plus de temps √† la progression
try:
    # Tentative de trouver la barre de progression
    try:
        progress_bar = WebDriverWait(driver, 10).until(
            EC.visibility_of_element_located((By.ID, "progress-bar"))
        )
        print("Barre de progression trouv√©e. Suivi de la progression...")
        
        # V√©rifier la progression jusqu'√† ce qu'elle atteigne 100%
        while True:
            # R√©cup√©rer la largeur actuelle de la barre de progression (en pourcentage)
            width = progress_bar.get_attribute("style")
            print("Largeur actuelle de la barre de progression : ", width)
            
            # Extraire la valeur du pourcentage de la largeur
            if "%" in width:
                percentage = float(width.split(":")[1].replace("%", "").strip())
                
                if percentage >= 100:  # Lorsque la progression atteint 100%
                    print("La progression a atteint 100%. Cliquer sur le bouton de t√©l√©chargement.")
                    break
            time.sleep(1)  # Attendre 1 seconde avant de v√©rifier √† nouveau
    except Exception as e:
        print("Barre de progression non trouv√©e, v√©rification du lien de t√©l√©chargement.")
    
    # Attendre que l'√©l√©ment <a> contenant l'URL soit visible
    try:
        # Utilisation d'un XPath pour trouver l'√©l√©ment <a> avec l'attribut href
        download_link = WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.XPATH, "//a[contains(@href, 'oceansaver.in')]"))
        )
    
        # Extraire l'attribut href de l'√©l√©ment trouv√©
        href = download_link.get_attribute("href")
        print(f"Lien trouv√© : {href}")
    
        # Cliquer sur le lien de t√©l√©chargement
        download_link.click()
    
        # Optionnel : Attendez un moment pour voir si le t√©l√©chargement commence
        time.sleep(2)
    
    except Exception as e:
        print(f"Erreur pendant l'interaction avec le lien de t√©l√©chargement : {e}")
        
except Exception as e:
    print(f"Erreur pendant l'attente ou le t√©l√©chargement : {e}")

# Fermer le navigateur apr√®s usage
driver.quit()
```

```python
import logging
from telegram import Update
from telegram.ext import Application, CommandHandler, MessageHandler, ConversationHandler, CallbackContext, filters
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select
import time

# Configurer le logger pour les erreurs
logging.basicConfig(format='%(asctime)s - %(name)s - %(levelname)s - %(message)s', level=logging.INFO)
logger = logging.getLogger(__name__)

# √âtats de la conversation
URL, FORMAT = range(2)

# Fonction de d√©marrage du bot
async def start(update: Update, context: CallbackContext) -> int:
    await update.message.reply_text("Bienvenue ! Veuillez entrer l'URL vid√©o (par exemple, YouTube) que vous souhaitez t√©l√©charger.")
    return URL

# Fonction pour recevoir l'URL de l'utilisateur
async def receive_url(update: Update, context: CallbackContext) -> int:
    context.user_data['url'] = update.message.text  # Sauvegarde l'URL
    await update.message.reply_text("Merci ! Choisissez un format parmi les options suivantes :\n"
                                    "1. 4k\n"
                                    "2. 1080p\n"
                                    "3. 720p\n"
                                    "Veuillez r√©pondre avec le num√©ro du format souhait√©.")
    return FORMAT

# Fonction pour recevoir le format de l'utilisateur
async def receive_format(update: Update, context: CallbackContext) -> int:
    format_choice = update.message.text.strip()

    # V√©rifier que le format est valide
    if format_choice == "1":
        selected_format = "4k"
    elif format_choice == "2":
        selected_format = "1080"
    elif format_choice == "3":
        selected_format = "720"
    else:
        await update.message.reply_text("Format invalide. Veuillez choisir un format parmi 1, 2 ou 3.")
        return FORMAT

    # Sauvegarder le format choisi
    context.user_data['format'] = selected_format
    await update.message.reply_text(f"Vous avez choisi le format {selected_format}. Je vais maintenant proc√©der au t√©l√©chargement.")

    # D√©marrer le processus de t√©l√©chargement avec Selenium
    download_video(context.user_data['url'], context.user_data['format'])

    await update.message.reply_text("T√©l√©chargement termin√© ! Si vous souhaitez recommencer, tapez /start.")
    return ConversationHandler.END

# Fonction pour t√©l√©charger la vid√©o via Selenium
def download_video(url, format_choice):
    driver = webdriver.Chrome()  # Assurez-vous que ChromeDriver est install√© et dans votre PATH
    driver.get("https://y2down.cc/fr/")  # Remplacez par l'URL de t√©l√©chargement r√©elle

    # Interagir avec l'input de l'URL
    input_element = WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.ID, "link"))
    )
    input_element.send_keys(url)

    # S√©lectionner le format vid√©o
    select_element = WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.ID, "format"))
    )
    select = Select(select_element)
    select.select_by_value(format_choice)  # S√©lectionner le format choisi

    # Cliquer sur le bouton de t√©l√©chargement
    download_button = driver.find_element(By.ID, "load")
    download_button.click()

    # Attendre le lien de t√©l√©chargement
    try:
        download_link = WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.XPATH, "//a[contains(@href, 'oceansaver.in')]"))
        )
        href = download_link.get_attribute("href")
        print(f"Lien de t√©l√©chargement : {href}")
        # Vous pouvez √©ventuellement ouvrir le lien pour d√©marrer le t√©l√©chargement
        download_link.click()
    except Exception as e:
        print(f"Erreur pendant l'interaction avec le lien de t√©l√©chargement : {e}")

    driver.quit()

# Fonction de gestion des erreurs
def error(update: Update, context: CallbackContext):
    logger.warning('Update "%s" caused error "%s"', update, context.error)

# Configuration de la conversation
conversation_handler = ConversationHandler(
    entry_points=[CommandHandler('start', start)],
    states={
        URL: [MessageHandler(filters.TEXT & ~filters.COMMAND, receive_url)],
        FORMAT: [MessageHandler(filters.TEXT & ~filters.COMMAND, receive_format)],
    },
    fallbacks=[],
)

async def main():
    # Remplacez par votre token d'API Telegram
    application = Application.builder().token("7753166825:AAGvSKCNiYKG1AHkkynIuNtqAvOmIOHdn2w").build()

    # Ajouter un gestionnaire de conversation
    application.add_handler(conversation_handler)

    # D√©marrer le bot
    await application.run_polling()

if __name__ == "__main__":
    # Au lieu d'utiliser asyncio.run, on appelle directement main
    import asyncio
    asyncio.create_task(main())  # Utilise create_task au lieu de run pour √©viter l'erreur
```

```python
!pip install nest_asyncio
import nest_asyncio
nest_asyncio.apply()

# le reste de ton code
import os
from pytube import YouTube
from telegram import Update
from telegram.ext import ApplicationBuilder, MessageHandler, CommandHandler, ContextTypes, filters

TOKEN = "7753166825:AAGvSKCNiYKG1AHkkynIuNtqAvOmIOHdn2w"
DOWNLOAD_FOLDER = "./videos"
os.makedirs(DOWNLOAD_FOLDER, exist_ok=True)

async def start(update: Update, context: ContextTypes.DEFAULT_TYPE):
    await update.message.reply_text("üëã Envoie-moi un lien YouTube.")


def is_youtube_url(text):
    return "youtube.com" in text or "youtu.be" in text

async def download_youtube(update: Update, context: ContextTypes.DEFAULT_TYPE):
    url = update.message.text
    if not is_youtube_url(url):
        await update.message.reply_text("‚ùå Lien non valide.")
        return
    try:
        yt = YouTube(url)
        stream = yt.streams.filter(file_extension='mp4', progressive=True).get_highest_resolution()
        stream.download(DOWNLOAD_FOLDER)
        await update.message.reply_text(f"‚úÖ T√©l√©charg√© : {yt.title}")
    except Exception as e:
        await update.message.reply_text(f"‚ùå Erreur : {str(e)}")

def main():
    app = ApplicationBuilder().token(TOKEN).build()
    app.add_handler(CommandHandler("start", start))
    app.add_handler(MessageHandler(filters.TEXT & (~filters.COMMAND), download_youtube))
    print("ü§ñ Bot lanc√©.")
    app.run_polling()

if __name__ == "__main__":
    main()
```
