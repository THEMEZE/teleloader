```python
import logging
import os
import nest_asyncio
import yt_dlp
import time

from telegram import Update
from telegram.ext import (
    Application,
    CommandHandler,
    MessageHandler,
    ConversationHandler,
    ContextTypes,
    filters
)

nest_asyncio.apply()

# ---------------------------------------
# CONFIG
# ---------------------------------------

TOKEN = "7753166825:AAGvSKCNiYKG1AHkkynIuNtqAvOmIOHdn2w"
DOWNLOAD_FOLDER = "./videos"
DOWNLOAD_FOLDER = "./Downloads"
DOWNLOAD_FOLDER = "/Users/themezeguillaume/Downloads"
os.makedirs(DOWNLOAD_FOLDER, exist_ok=True)

# Logger
logging.basicConfig(format='%(asctime)s - %(name)s - %(levelname)s - %(message)s', level=logging.INFO)
logger = logging.getLogger(__name__)

# Ã‰tats de la conversation
URL, FORMAT = range(2)

# ---------------------------------------
# FONCTIONS yt-dlp
# ---------------------------------------

def download_with_ytdlp(url: str, format_choice: str) -> str:
    """
    TÃ©lÃ©charge la vidÃ©o via yt-dlp et retourne le chemin du fichier.
    """

    # Options gÃ©nÃ©rales
    ydl_opts = {
        'outtmpl': os.path.join(DOWNLOAD_FOLDER, '%(title)s.%(ext)s'),
        'noplaylist': True,
        'quiet': True,
        'no_warnings': True
    }

    # Choix du format
    if format_choice == "mp3":
        ydl_opts.update({
            'format': 'bestaudio/best',
            'postprocessors': [{
                'key': 'FFmpegExtractAudio',
                'preferredcodec': 'mp3',
                'preferredquality': '192',
            }]
        })
    else:
        ydl_opts['format'] = 'bestvideo+bestaudio/best'

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        info = ydl.extract_info(url, download=True)

    # RÃ©cupÃ©ration du fichier tÃ©lÃ©chargÃ©
    filename = ydl.prepare_filename(info)

    if format_choice == "mp3":
        filename = filename.rsplit(".", 1)[0] + ".mp3"

    return filename

# ---------------------------------------
# BOT TELEGRAM
# ---------------------------------------

async def start(update: Update, context: ContextTypes.DEFAULT_TYPE) -> int:
    await update.message.reply_text("ğŸ‘‹ Envoie-moi une URL YouTube / Instagram / TikTok / etc.")
    return URL

async def receive_url(update: Update, context: ContextTypes.DEFAULT_TYPE) -> int:
    context.user_data['url'] = update.message.text.strip()

    await update.message.reply_text(
        "ğŸ¯ Choisis un format :\n"
        "1ï¸âƒ£ VidÃ©o MP4 (meilleure qualitÃ©)\n"
        "2ï¸âƒ£ Audio MP3"
    )
    return FORMAT

async def receive_format(update: Update, context: ContextTypes.DEFAULT_TYPE) -> int:
    choice = update.message.text.strip()

    if choice == "1":
        format_choice = "mp4"
    elif choice == "2":
        format_choice = "mp3"
    else:
        await update.message.reply_text("âŒ Choix invalide. Tape 1 ou 2.")
        return FORMAT

    url = context.user_data['url']
    await update.message.reply_text("â³ TÃ©lÃ©chargement en cours...")

    try:
        file_path = download_with_ytdlp(url, format_choice)

        await update.message.reply_text("ğŸ“¤ Envoi du fichier...")

        with open(file_path, "rb") as f:
            await update.message.reply_document(f)

        await update.message.reply_text("âœ… TerminÃ© ! Tape /start pour recommencer.")

    except Exception as e:
        await update.message.reply_text(f"âŒ Erreur : {e}")
        logger.error(e)

    return ConversationHandler.END


def main():
    app = Application.builder().token(TOKEN).build()

    conv = ConversationHandler(
        entry_points=[CommandHandler("start", start)],
        states={
            URL: [MessageHandler(filters.TEXT & ~filters.COMMAND, receive_url)],
            FORMAT: [MessageHandler(filters.TEXT & ~filters.COMMAND, receive_format)]
        },
        fallbacks=[]
    )

    app.add_handler(conv)

    print("ğŸ¤– Bot lancÃ©.")
    app.run_polling()


if __name__ == "__main__":
    main()
```

```python
import logging
import os
import nest_asyncio
import yt_dlp
import time

from telegram import Update
from telegram.ext import (
    Application,
    CommandHandler,
    MessageHandler,
    ConversationHandler,
    ContextTypes,
    filters
)

nest_asyncio.apply()

# ---------------------------------------
# CONFIG
# ---------------------------------------

TOKEN = "7753166825:AAGvSKCNiYKG1AHkkynIuNtqAvOmIOHdn2w"
DOWNLOAD_FOLDER = "./videos"
DOWNLOAD_FOLDER = "./Downloads"
DOWNLOAD_FOLDER = "/Users/themezeguillaume/Downloads"
os.makedirs(DOWNLOAD_FOLDER, exist_ok=True)



# ---------------------------------------
# FONCTIONS yt-dlp
# ---------------------------------------

def download_with_ytdlp(url: str, format_choice: str) -> tuple[str, dict]:
    """
    TÃ©lÃ©charge la vidÃ©o via yt-dlp et retourne le chemin du fichier et les infos.
    """
    ydl_opts = {
        'outtmpl': os.path.join(DOWNLOAD_FOLDER, '%(title)s.%(ext)s'),
        'noplaylist': True,
        'quiet': True,
        'no_warnings': True
    }

    if format_choice == "mp3":
        ydl_opts.update({
            'format': 'bestaudio/best',
            'postprocessors': [{
                'key': 'FFmpegExtractAudio',
                'preferredcodec': 'mp3',
                'preferredquality': '192',
            }]
        })
    else:
        ydl_opts['format'] = 'bestvideo+bestaudio/best'

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        # Extraire les infos sans tÃ©lÃ©charger pour afficher Ã  l'utilisateur
        info = ydl.extract_info(url, download=False)
        title = info.get("title")
        uploader = info.get("uploader")
        duration = info.get("duration")

        # TÃ©lÃ©chargement rÃ©el
        ydl.download([url])

        filename = ydl.prepare_filename(info)
        if format_choice == "mp3":
            filename = filename.rsplit(".", 1)[0] + ".mp3"

    return filename, {"title": title, "uploader": uploader, "duration": duration}

# ---------------------------------------
# BOT TELEGRAM
# ---------------------------------------

# ---------------------------------------
# BOT TELEGRAM
# ---------------------------------------

async def start(update: Update, context: ContextTypes.DEFAULT_TYPE) -> int:
    await update.message.reply_text("ğŸ‘‹ Envoie-moi une URL YouTube / Instagram / TikTok / etc.")
    return URL

async def receive_url(update: Update, context: ContextTypes.DEFAULT_TYPE) -> int:
    context.user_data['url'] = update.message.text.strip()

    await update.message.reply_text(
        "ğŸ¯ Choisis un format :\n"
        "1ï¸âƒ£ VidÃ©o MP4 (meilleure qualitÃ©)\n"
        "2ï¸âƒ£ Audio MP3"
    )
    return FORMAT

async def receive_format(update: Update, context: ContextTypes.DEFAULT_TYPE) -> int:
    choice = update.message.text.strip()

    if choice == "1":
        format_choice = "mp4"
    elif choice == "2":
        format_choice = "mp3"
    else:
        await update.message.reply_text("âŒ Choix invalide. Tape 1 ou 2.")
        return FORMAT

    url = context.user_data['url']
    await update.message.reply_text("â³ TÃ©lÃ©chargement en cours...")

    try:
        file_path, info = download_with_ytdlp(url, format_choice)

        print("file_path, info" , file_path, info) 

        # Affichage des infos dans Telegram
        await update.message.reply_text(
            f"ğŸ“¹ Titre : {info['title']}\n"
            f"ğŸ“º Uploader : {info['uploader']}\n"
            f"â±ï¸ DurÃ©e : {info['duration']} sec"
        )

        await update.message.reply_text("ğŸ“¤ Envoi du fichier...")

        with open(file_path, "rb") as f:
            await update.message.reply_document(f)

        await update.message.reply_text("âœ… TerminÃ© ! Tape /start pour recommencer.")

    except Exception as e:
        await update.message.reply_text(f"âŒ Erreur : {e}")
        logger.error(e)

    return ConversationHandler.END


def main():
    app = Application.builder().token(TOKEN).build()

    conv = ConversationHandler(
        entry_points=[CommandHandler("start", start)],
        states={
            URL: [MessageHandler(filters.TEXT & ~filters.COMMAND, receive_url)],
            FORMAT: [MessageHandler(filters.TEXT & ~filters.COMMAND, receive_format)]
        },
        fallbacks=[]
    )

    app.add_handler(conv)

    print("ğŸ¤– Bot lancÃ©.")
    app.run_polling()


if __name__ == "__main__":
    main()
```
